```python
import xgboost
import numpy as np
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier, plot_tree
from sklearn.metrics import classification_report

import matplotlib.pyplot as plt
```

```python
dataset=np.loadtxt('data_banknote_authentication.txt', delimiter=',', dtype=str, )
dataset=[[float(n) for n in e] for e in dataset]
X=[]
y=[]
for row in dataset:
    X.append(row[:-1])
    y.append(int(row[-1]))
X=np.array(X)
y=np.array(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.5, random_state=0)

scores = ['precision', 'recall', 'f1', 'accuracy', 'roc_auc']
```

```python
tuned_parameters = [{
    'n_estimators': [1, 2, 3, 9, 12, 15, 30, 50, 100, 150, 300, 500, 1000],
    'learning_rate': [0.00000000001, 0.0000000001, 0.000000001, 0.00000001, 0.0000001, 0.000001],
    'max_depth': [3, 5, 6, 9, 12]
}]

for score in scores:
    print("# Tuning hyper-parameters for %s" % score)
    print()

    clf = GridSearchCV(XGBClassifier(), tuned_parameters,
                       scoring = score, cv=2, refit='auc_roc')
    clf.fit(X=X_train,y=y_train)

    print("Best parameters set found on development set:")
    print()
    print(clf.best_params_)
    print()
    print("Grid scores on development set:")
    print()
    means = clf.cv_results_['mean_test_score']
    stds = clf.cv_results_['std_test_score']
    for mean, std, params in zip(means, stds, clf.cv_results_['params']):
        print("%0.3f (+/-%0.03f) for %r"
              % (mean, std * 2, params))
    print()

    print("Detailed classification report:")
    print()
    print("The model is trained on the full development set.")
    print("The scores are computed on the full evaluation set.")
    print()
    y_true, y_pred = y_test, clf.predict(X_test)
    print(classification_report(y_true, y_pred))
    print()
```

```python
model=XGBClassifier(n_estimators=9,learning_rate=0.0000001,max_depth=9)
model.fit(X,y)
plot_tree(model)
plt.show()
```
